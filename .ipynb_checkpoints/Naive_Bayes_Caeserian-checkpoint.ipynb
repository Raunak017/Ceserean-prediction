```python
# Loading Dataset
import pandas as pd

data = pd.read_csv ('./caeserian.csv')
print(data)
```

```python
#No Need for Label Encoding as Data is already encoded

data2 = data.copy()
data2 = data.drop(['Caesarian'], axis=1)

X = data2.iloc[:,0:].values
y = data.iloc[:, 5].values
```

```python
# Splitting data 70%-30% into training set and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)

print(X_train)

print(y_train)

print(X_test)

print(y_test)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train)

print(X_test)
```

```python
#Training Model on the Training set
import numpy as np
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)

#Predicting the Test set results
y_pred = model.predict(X_test)
print('Predicted Labels: ',y_pred)
print('Actual Labels:    ',y_test)                            
```

```python
# Predicting Output for Custom Data
# {"Age":'29', "Delivery Number":"3", "Delivery Time":'Premature' , "Blood of Pressure":'Normal', "Heart Problem":'inept'}
# The data is encoded as [29,3,1,1,1]
A = np.array([29,3,1,1,1])
prediction = model.predict(A.reshape(1,-1))
print('Prediction for Classified Data',A ,':',prediction)

# {"Age":'21', "Delivery Number":"2", "Delivery Time":'Timely' , "Blood of Pressure":'High', "Heart Problem":'apt'}
# The data is encoded as [21,2,0,2,0]
B = np.array([21,2,0,2,0])
prediction = model.predict(B.reshape(1,-1))
print('Prediction for Classified Data',B ,':',prediction)
```

```python
#Confusion Matrix and Accuracy

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix: ',cm)
print('Accuacy Score:    ',accuracy_score(y_test, y_pred))
print('Precision Score:  ',precision_score(y_test, y_pred))
print('Recall Score:     ',recall_score(y_test, y_pred))
print('F-Score:          ',f1_score(y_test, y_pred))
```
